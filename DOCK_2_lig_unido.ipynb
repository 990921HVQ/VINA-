```python
import py3Dmol

def definegrid(object,bxi,byi,bzi,bxf,byf,bzf):
  object.addBox({'center':{'x':bxi,'y':byi,'z':bzi},'dimensions': {'w':bxf,'h':byf,'d':bzf},'color':'blue','opacity': 0.6})

def viewprot(object,prot_PDBfile,resids):
  mol1 = open(prot_PDBfile, 'r').read()
  object.addModel(mol1,'pdb')
 

  ChA={'chain':'A'}
  ChB={'chain':'B'}
  BAH1={'resi':'755-880'}
  BAH2={'resi':'972-1100'}
  Autoinhibitor_linker={'resi':'693-754'}
  Binding_site={'resi':'1146-1191'}
  Active_site={'resi':'1226'}


  object.addSurface(py3Dmol.VDW, {'color':'spectrum', 'opacity':0.85}, Binding_site)

  object.setStyle(ChA, {'cartoon':{'color':'lightblue', 'arrows':True}})
  object.setStyle(ChB,{'cartoon':{'color':'darkorange', 'arrows':True}})
  object.setStyle(BAH1, {'cartoon':{'color':'green','arrows':True}}) 
  object.setStyle(BAH2, {'cartoon':{'color':'purple', 'arrows':True}})
  object.setStyle(Autoinhibitor_linker, {'cartoon':{'color':'red', 'arrows':True}})
  object.setStyle(Binding_site, {'cartoon':{'color':'cyan', 'arrows':True}})
  object.setStyle(Active_site, {'cartoon':{'color':'yellow', 'arrows':True}})


  object.addStyle({'resi':resids},{'stick':{'colorscheme':'greenCarbon'}})


def viewprotgrid(prot_PDBfile,resids,bxi,byi,bzi,bxf=10,byf=10,bzf=10):
  mol_view = py3Dmol.view(1000,600)
  definegrid(mol_view,bxi,byi,bzi,bxf,byf,bzf)
  viewprot(mol_view,prot_PDBfile,resids)
  mol_view.setBackgroundColor('0xffffff')
  mol_view.zoomTo()
  mol_view.show()
```

```python
from ipywidgets import interact,fixed,IntSlider
import ipywidgets
interact(viewprotgrid,
         prot_PDBfile = ['4wxx_cadena_b_preparada.pdb'],
#Agrega los residuos que quieres visualizar
         resids = [(1145,1146, 1151, 1168, 1169, 1190, 1191, 1578, 1580)],
         bxi=ipywidgets.IntSlider(min=-100,max=100, step=1),
         byi=ipywidgets.IntSlider(min=-100,max=100, step=1),
         bzi=ipywidgets.IntSlider(min=-100,max=100, step=1),
         bxf=ipywidgets.IntSlider(min=0,max=100, step=1),
         byf=ipywidgets.IntSlider(min=0,max=100, step=1),
         bzf=ipywidgets.IntSlider(min=0,max=100, step=1))
```
